```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import scipy.interpolate
from mpl_toolkits.axes_grid1 import make_axes_locatable
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import
from scipy.signal import find_peaks

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

plt.rcParams.update({'font.size': 14,
                     'axes.titlesize': 18,
                     'axes.labelsize': 16,
                     'axes.labelpad': 14,
                     'lines.linewidth': 1,
                     'lines.markersize': 10,
                     'xtick.labelsize' : 16,
                     'ytick.labelsize' : 16,
                     'xtick.top' : True,
                     'xtick.direction' : 'in',
                     'ytick.right' : True,
                     'ytick.direction' : 'in',})
```

## Data for the tube with different length

```python
freq6,amp6=np.loadtxt('6_units_without_aperture/6_units_without_aperture',unpack=True)
peaks6, _ = find_peaks(amp6, distance=30)
freq7,amp7=np.loadtxt('7_units_without_aperture/7_units_without_aperture',unpack=True)
peaks7, _ = find_peaks(amp7, distance=30)
freq8,amp8=np.loadtxt('8_units_without_aperture/8_units_without_aperture',unpack=True)
peaks8, _ = find_peaks(amp8, distance=30)
```

```python
fig=plt.figure(figsize=(10,8))
ax=fig.gca()
plt.subplot(311)
plt.plot(freq6,amp6/10000,label='L=30 cm')
plt.plot(freq6[peaks6],amp6[peaks6]/10000,'.')
plt.ylabel('amplitude [a.u.]')
plt.legend()
plt.subplot(312)
plt.plot(freq7,amp7/10000,label='L=35 cm')
plt.plot(freq7[peaks7],amp7[peaks7]/10000,'.')
plt.ylabel('amplitude [a.u.]')
plt.legend()
plt.subplot(313)
plt.plot(freq8,amp8/10000,label='L=40 cm')
plt.plot(freq8[peaks8],amp8[peaks8]/10000,'.')
plt.legend()
plt.xlabel('frequency [Hz]')
plt.ylabel('amplitude [a.u.]')
plt.ticklabel_format(axis='both',style='sci')
plt.tight_layout()
plt.savefig('resonances_wo_apertures.pdf')
plt.show()
```

```python
k6=np.arange(1,19)*np.pi/(30)
plt.plot(k6,freq6[peaks6[1:]]/1000,'ro',alpha=0.3,label='L=30 cm')
k7=np.arange(1,22)*np.pi/(35)
plt.plot(k7,freq7[peaks7[1:]]/1000,'go',alpha=0.3,label='L=35 cm')
k8=np.arange(1,25)*np.pi/(40)
plt.plot(k8,freq8[peaks8[1:]]/1000,'bo',alpha=0.3,label='L=40 cm')
plt.xlabel('wavenumber k [cm$^{-1}$]')
plt.ylabel('frequency [kHz]')
plt.legend()
plt.tight_layout()
plt.savefig('k_for_all.pdf')
plt.show()
```
